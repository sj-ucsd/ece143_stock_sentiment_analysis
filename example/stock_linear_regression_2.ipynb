```python
import yahooquery
import numpy as np
import pandas as pd
import matplotlib as mp
```

```python
symbol = 'AAPL'
ticker = yahooquery.Ticker(symbol)
```

```python
# get price history - default is year to date but can specify period and frequency
hist = ticker.history()
```

```python
hist = pd.DataFrame(hist)
```

```python
#drop columns we don't plan to use from the history
hist = hist.drop(['high', 'open', 'low' , 'adjclose'], axis=1)
```

```python
# We would like to use previus day's price and volume as inputs to predict today's price hence 
# shift values by one to create new columns 
hist['previous_close'] = hist['close'].shift(-1)
hist['previous_volume'] = hist['volume'].shift(-1)
```

```python
hist
```

```python
# drop any rows with data that is unavailable (NaN)
hist=hist.dropna()
```

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn import dummy, metrics, model_selection, preprocessing
from sklearn import model_selection
```

```python
# X is the input and Y is the output. 
x_columns = ['previous_close', 'previous_volume']
y_columns = ['close']
```

```python
# use sklearn to split the data automatically into training and testing data. 30% is for test in this case but can be changed
X_train, X_test, y_train, y_test = model_selection.train_test_split(hist[x_columns],hist[y_columns]\
                                                                                  , test_size=0.3, random_state=42)
cols = X_train.columns
std = preprocessing.StandardScaler()
X_train.loc[:, cols] = std.fit_transform(X_train)
X_test.loc[:,cols] = std.transform(X_test)
```

```python
# instantiate linear regression model and fit  using training data
lr = LinearRegression()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# score using test data
lr.score(X_test, y_test)
```
